```python
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np
import os
```

```python
df2 = pd.read_csv(r'/Users/philip/Documents/TUBerlin/BA/FQBench_Maven/my-app/src/main/out/TD1_SF1/FQBenchResult.csv')
df2.sort_values("Query", axis=0, ascending=True,
                 inplace=True, na_position='first')
df2 = df2.pivot(index='Query', columns='FederatedEngine', values='Runtime')
df2.plot(kind='bar', figsize=(15, 8), color=['green', 'red', 'blue'])
plt.ylim((0,60000))
plt.xlabel('Query')
plt.ylabel('Runtime in ms')
plt.title('Query Runtimes per Federated Query Engine (TD1)')
plt.show()
```
